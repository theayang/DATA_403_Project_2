```python
import pandas as pd
df = pd.read_csv('/Users/nickhausman/Desktop/DATA/P2/noncorrelated_train.csv')
df.head()
```

```python
cols = 'APARTMENTS_MEDI	BASEMENTAREA_MEDI	YEARS_BEGINEXPLUATATION_MEDI	YEARS_BUILD_MEDI	COMMONAREA_MEDI	ENTRANCES_MEDI	FLOORSMAX_MEDI	FLOORSMIN_MEDI	LANDAREA_MEDI	NONLIVINGAPARTMENTS_MEDI	NONLIVINGAREA_MEDI'.split()
combine = df[cols]
len(cols)
```

```python
import numpy as np
```

```python
housing_columns_above_mean_cnt = pd.Series(np.zeros(len(combine.index)))
for col in combine.columns:
    housing_columns_above_mean_cnt += (combine[col] > combine[col].mean()).astype(int)
housing_columns_above_mean_cnt.value_counts()
```

```python
na_bool_series = [combine[col].isna() for col in combine.columns]
print(na_bool_series)
undefined_housing_indicies = []
for i in range(len(na_bool_series[0])):
    if all(l[i] for l in na_bool_series):
        undefined_housing_indicies.append(i)
len(undefined_housing_indicies) / len(combine.index)
```

```python
for i in undefined_housing_indicies: housing_columns_above_mean_cnt[i] = 'No Housing information' 
housing_columns_above_mean_cnt.value_counts()
```

```python
df['housing_columns_above_mean_cnt'.upper()] = housing_columns_above_mean_cnt
df.to_csv('After_housing_modifications.csv')
df.columns
```
